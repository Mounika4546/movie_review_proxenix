```python
import pandas as pd
import matplotlib.pyplot as plt
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
data=pd.read_csv('review.csv');
```

```python
data['sentiment'].value_counts()
```

```python
data.dropna(inplace=True)
```

```python
def clean_review(review):
    str=' '.join(word for word in review.split() if word.lower() not in stopwords.words('english'))
    return str
```

```python
data['revi']=data['text'].apply(clean_review)
```

```python
print(data.columns)
# Inspect sentiment labels first
print(data['sentiment'].unique())
```

```python
data.rename(columns={'review': 'text'}, inplace=True)
```

```python
#wordcloud

from wordcloud import WordCloud
```

```python
#draw wordcloud
reviews=' '.join(word for word in data['text'][data['sentiment']=='positive'].astype(str))
wordcloud=WordCloud(height=600,width=1000,max_font_size=100)
plt.figure(figsize=(15,12))
plt.imshow(wordcloud.generate(reviews),interpolation='bilinear')
plt.axis('off');
plt.show()
```

```python
reviews=' '.join(word for word in data['text'][data['sentiment']=='negative'].astype(str))
wordcloud=WordCloud(height=600,width=1000,max_font_size=100)
plt.figure(figsize=(15,12))
plt.imshow(wordcloud.generate(reviews),interpolation='bilinear')
plt.axis('off');
plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv=TfidfVectorizer(max_features=2500)
```

```python
reviews=cv.fit_transform(data['text']).toarray()
```

```python
data['sentiment']=data['sentiment'].replace(['positive','negative'],[1,0]).astype(int)
```

```python
data['sentiment'].value_counts()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

from sklearn.model_selection import train_test_split

```python
from sklearn.model_selection import train_test_split
```

```python
reviews_train,reviews_test,sent_train,sent_test=train_test_split(reviews,data['sentiment'],test_size=0.2)
```

```python
model.fit(reviews_train,sent_train)
```

```python
predict=model.predict(reviews_test)
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cm=confusion_matrix(sent_test,predict,labels=model.classes_)
display=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=model.classes_)
display.plot()
plt.show()
```
